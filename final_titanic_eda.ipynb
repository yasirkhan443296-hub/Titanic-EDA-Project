# **Step 5: Missing Value Treatment

**Columns with missing values:**
- Age: ~20% missing → IterativeImputer
- Fare: 1 missing → median fill
- Embarked: 2 missing → mode fill
- Cabin: ~77% missing → drop or mark Unknown

```python
# Save original Age before filling (for comparison)
age_before = df["Age"].copy()

# IterativeImputer — uses Pclass, Fare, SibSp, Parch to predict Age
imputer = IterativeImputer(max_iter=10, random_state=42)
df["Age"] = pd.Series(
    imputer.fit_transform(
        df[["Age","Pclass","Fare","SibSp","Parch"]]
    )[:,0],
    index=df.index
)
df["Age"] = df["Age"].astype("float64")
df["Age"] = df["Age"].clip(lower=0)
print("Age nulls after:", df["Age"].isnull().sum())
```

```python
# Compare Age before vs after — KDE plot
age_before.plot(kind="kde", label="Before (with nulls)",
                color="red", linestyle="dashed")
df["Age"].plot(kind="kde", label="After (filled)",
               color="green")
plt.title("Age: Before vs After IterativeImputer")
plt.xlabel("Age")
plt.legend()
plt.show()
```

```python
# Show which rows were filled
null_index = df[age_before.isnull()].index
comparison = pd.DataFrame({"Before": age_before, "After": df["Age"]})
print("Total filled:", len(null_index))
comparison[age_before.isnull()].head(10)
```

```python
# Fill Fare — 1 missing value → median
df["Fare"].fillna(df["Fare"].median(), inplace=True)
print("Fare nulls after:", df["Fare"].isnull().sum())
```

```python
# Fill Embarked — 2 missing → mode (most common = S)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
print("Embarked nulls after:", df["Embarked"].isnull().sum())
```

```python
# Cabin — 77% missing → fill with Unknown
df["Cabin"].fillna("Unknown", inplace=True)
print("Cabin nulls after:", df["Cabin"].isnull().sum())
```

```python
# Final verification
print("=== Missing Values After Treatment ===")
print(df.isnull().sum())
```

**CONCLUSION — Missing Value Treatment**

° Age (20% missing) → filled using IterativeImputer with Pclass, Fare, SibSp, Parch — most accurate method

° Fare (1 missing) → filled with median — safe for very small missing %

° Embarked (2 missing) → filled with mode (S = Southampton) — standard for categorical

° Cabin (77% missing) → filled with 'Unknown' — too high to impute reliably

° KDE plot confirms Age distribution was well preserved after imputation ✅

# **Step 6: Feature Engineering

```python
# Investigate group tickets (Sage family)
df[df["SibSp"]==8]
```

```python
df[df["Ticket"]=="CA. 2343"]
```

```python
# individual_fare = Fare per person
df["individual_fare"] = df["Fare"] / (df["SibSp"] + df["Parch"] + 1)
df[["individual_fare","Fare"]].describe()
```

```python
df["individual_fare"].plot(kind="box")
plt.title("Individual Fare Distribution")
plt.show()
```

```python
df[df["individual_fare"]>490]
```

```python
# family_size
df["family_size"] = df["SibSp"] + df["Parch"] + 1
df["family_size"].value_counts()
```

```python
# family_type
def transform_family_size(nums):
    if nums == 1:
        return "alone"
    elif nums > 1 and nums < 5:
        return "small"
    else:
        return "large"

df["family_type"] = df["family_size"].apply(transform_family_size)
pd.crosstab(df["family_type"], df["Survived"], normalize="columns")*100
```

```python
# Extract surname and title from Name
df["surname"] = df["Name"].str.split(",").str.get(0)
df["title"]   = df["Name"].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0)
df[["Name","surname","title"]].head(10)
```

**CONCLUSION — Feature Engineering**

° **individual_fare**: Created by dividing Fare by group size — gives accurate fare per person

° **family_size**: SibSp + Parch + 1 — ranges from 1 (alone) to 11 (largest family)

° **family_type**: Alone (1) → Small (2–4) → Large (5+)

° Crosstab shows small families had the best survival rate

° **title** and **surname** extracted from Name — useful for grouping by social status
